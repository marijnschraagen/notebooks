# Combine results of keyword frequency counts

```python
import json
import pandas as pd

BASEDIR = "../keywords/./"
INFILENAME1 = "keywords.out.20200424"
TOTAL = "TOTAL"

def changeTokens(dataIn):
    dataOut = {}
    for token in dataIn:
        dataOut["*"+token+"*"] = dataIn[token]
    return(dataOut)

def addTotals(data):
    for token in data:
        data[token][TOTAL] = sum(data[token].values())
    return(data)
    
def readFile(inFileName):
    inFile = open(inFileName,"r")
    data = {}
    for line in inFile:
        jsonData = json.loads(line)
        for token in jsonData:
            data[token] = jsonData[token]
    return(data)

data = changeTokens(addTotals(readFile(BASEDIR+INFILENAME)))
sortedData = {k:v for k,v in sorted(data.items(), key=lambda item: item[1][TOTAL], reverse=True)}
df = pd.DataFrame.from_dict(sortedData).T
df.to_csv("keywords2.csv",index_label="keyword")
df
```

```python
INFILENAME2 = "keywords.out.20200423"
INFILENAME3 = "keywords.out.20200420"
INFILENAME4 = "keywords.out.covid"

def combineDicts(dict1,dict2):
    combinedDict  = {}
    for keyword in dict1:
        if not keyword in dict2: 
            combinedDict[keyword] = dict1[keyword]
        else:
            combinedDict[keyword] = combineDicts(dict1[keyword],dict2[keyword])
    for keyword in dict2:
        if not keyword in dict1: 
            combinedDict[keyword] = dict2[keyword]
    return(combinedDict)
            
data = {}
for inFileName in [INFILENAME2,INFILENAME3,INFILENAME4]:
    dataPart = readFile(BASEDIR+inFileName)
    data = combineDicts(dataPart,data)
data = addTotals(data)
sortedData = {k:v for k,v in sorted(data.items(), key=lambda item: item[1][TOTAL], reverse=True)}
df = pd.DataFrame.from_dict(sortedData).T
df.to_csv("keywords1.csv",index_label="keyword")
df
```
